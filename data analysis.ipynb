```python
# Import Titanic dataset (available in seaborn 'or
import seaborn as sns 
df = sns.load_dataset("titanic")

# Inspect 
print(df.shape)
print(df.head())
print(df.info())
print(df.describe())
```

```python
df['age'] = df['age'].fillna(df['age'].median())
print(df['age'])
```

```python
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
print(df['embarked'])
```

```python
df['family_size'] = df['sibsp'] + df['parch'] + 1
print(df['family_size'])
```

```python
overall_survival = df['survived'].mean()
```

```python
survival_by_class = df.groupby('pclass')['survived'].mean()
```

```python
survival_by_gender = df.groupby('sex')['survived'].mean()
```

```python
age_by_survival = df.groupby('survived')['age'].mean()
```

```python
overall_survival = df['survived'].mean()
survival_by_class = df.groupby('pclass')['survived'].mean()
survival_by_gender = df.groupby('sex')['survived'].mean()
age_by_survival = df.groupby('survived')['age'].mean()
print("overall survival rate:", overall_survival)
print("\nsurvival by class:\n", survival_by_class)
print("\nsurvival by gender:\n", survival_by_gender)
print("\naverage age by survival:\n", age_by_survival)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
sns.countplot(x='survived',data=df)
plt.title("overall survival count")
plt.show()
sns.barplot(x='pclass', y='survived', data=df)
plt.title("survival rate by class")
plt.show()
sns.barplot(x='sex', y='survived', data=df)
plt.title("survival rate by gender")
plt.show()
sns.scatterplot(x='age', y='fare', hue='survived' , data=df)
plt.title("age vs fare with survival")
plt.show()
sns.boxplot(x='pclass', y='age' , data=df)
plt.title("age distribution by passenger class")
plt.show()
```

overall survival rate = 38%
women survived more(74%)compared to men(19%)
class mattered: first class survival(63%)>second(47%)>third(24%)
age mattered: children had higher survival than older passengers
